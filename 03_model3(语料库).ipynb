```python
import pandas as pd

df1 = pd.read_csv("text.csv", encoding="utf-8-sig")
df1.set_index(["时间", "序号"], inplace=True)
df1.sort_index(inplace=True)
df1
```

```python
df2 = pd.read_csv("D:/Desktop/df.csv", encoding="gbk")
df2.set_index(["时间", "序号"], inplace=True)
df2.sort_index(inplace=True)
df2
```

```python
df_ = df2[df2.index.isin(df1.index)]
df_
```

```python
def accuracy(pred, y):
    acc = 0
    for i in range(len(y)):
        if pred[i] == y[i]:
            acc += 1
    return acc / len(pred)
```

```python
accuracy(df_["pred"].values, df1["情绪打分"].values)
```
