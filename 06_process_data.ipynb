```python
import pandas as pd

df = pd.read_csv("result.csv", encoding="gbk")
df["时间"] = pd.to_datetime(df["时间"])
df.set_index(["时间", "股票代码"], inplace=True)
df.sort_index(inplace=True)
df
```

```python
df.index.get_level_values(1).unique()
```

```python
price_data = pd.read_csv('tot2.csv')
price_data
```

将股票代码整理成更通用的格式

```python
price_data["code"] = price_data["code"].str.replace("XSHE", "SZ")
price_data["code"] = price_data["code"].str.replace("XSHG", "SH")
price_data
```

```python
price_data["time"] = pd.to_datetime(price_data["time"])
#price_data = price_data.set_index(["time", "code"]).sort_index()
#price_data["close"].groupby("time").count()
```

每天的新闻数据(条)

```python
df["predict"].groupby("时间").count()
```

```python
df = df[df.index.get_level_values(1).isin(price_data["code"])]
df["predict"].groupby("时间").count()
```

```python
count = pd.DataFrame(df["predict"].groupby("时间").count())
t_delete = count[count.values < 50].index.get_level_values(0)
t_delete
```

```python
df = df[~df.index.get_level_values(0).isin(t_delete.values)]
df["predict"].groupby("时间").count()
```

```python
sentiment_index = pd.DataFrame(df["predict"].groupby("时间").mean())
sentiment_index.describe()
```

```python
sentiment_index.index.names = ["time"]
sentiment_index.index
```

```python
price_data = price_data[price_data["time"].isin(sentiment_index.index.unique().values)]
price_data
```

```python
import scutquant.scutquant as q

price_data = price_data.set_index(["time", "code"]).sort_index()
data = q.join_data(price_data, sentiment_index, time="time", col=["predict"],index=["time", "code"])
data
```

```python
data.describe()
```

```python
data.describe().to_csv("D:/Desktop/describe.csv")
```

```python
data.sort_index(inplace=True)
data["close"].groupby("code").count()
```

```python
data["close"].groupby("time").count()
```

```python
data["ret"] = data["close"] / data["pre_close"] - 1
data["ret"].clip(-0.1, 0.1, inplace=True)
data["index"] = data["ret"] / data["ret"].groupby("time").mean() # 个股收益与大盘收益之比, 衡量个股相对大盘的强弱
data["label"] = data["ret"].groupby("code").shift(-2)
data["label"].clip(-0.1, 0.1, inplace=True)
data.dropna(inplace=True)
data
```
